```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/diwali-sales-dataset/Diwali Sales Data.csv')
```

```python
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='Gender', data=df, order=df['Gender'].value_counts().index, color='#124456')
plt.title('Gender Count')
plt.xlabel('Gender')
plt.ylabel('Count')

# Add labels on top of bars
for p in ax.patches:
    count = int(p.get_height())
    ax.text(p.get_x() + p.get_width()/2, count + 1, str(count), ha='center')

plt.show()
```

```python
df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sells_gen=df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sells_gen)
```

**most of the buyers are female and purchashing power of females are high**

```python
ax=sns.countplot(data=df,x = 'Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sells_age=df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sells_age)
```

**From the above graph we can tell this that most of the buyers are the between age 26 to 35**
